```python
img = cv2.imread("Desk.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h, w = img.shape[:2]
```

```python
# Rotation
angle = 45
M_rotation = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
rotated_img = cv2.warpAffine(img, M_rotation, (w, h))
```

```python
# Scaling
sx, sy = 1.5, 1.5
M_scaling = np.float32([[sx, 0, 0], [0, sy, 0]])
scaled_img = cv2.warpAffine(img, M_scaling, (int(w * sx), int(h * sy)))
```

```python
# Shearing
shx, shy = 0.5, 0.5
M_shearing = np.float32([[1, shx, 0], [shy, 1, 0]])
sheared_img = cv2.warpAffine(img, M_shearing, (int(w + shx * h), int(h + shy * w)))
```

```python
# Reflection
M_reflection = np.float32([[-1, 0, w], [0, 1, 0]])
reflected_img = cv2.warpAffine(img, M_reflection, (w, h))
```

```python

# Display all transformations
fig, axs = plt.subplots(1, 5, figsize=(20, 10))
axs[0].imshow(translated_img)
axs[0].axis('off')
axs[0].set_title('Translation')

axs[1].imshow(rotated_img)
axs[1].axis('off')
axs[1].set_title('Rotation')

axs[2].imshow(scaled_img)
axs[2].axis('off')
axs[2].set_title('Scaling')

axs[3].imshow(sheared_img)
axs[3].axis('off')
axs[3].set_title('Shearing')

axs[4].imshow(reflected_img)
axs[4].axis('off')
axs[4].set_title('Reflection')

plt.show()
```
